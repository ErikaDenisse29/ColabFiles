# **4. Limpieza y Transformación del Conjunto de Datos**

 **Normalización de Nombres de Columnas:**

En este dataset algunos títulos utilizan mayúsculas al inicio de cada palabra, mientras que otros tienen guiones bajos. Se cambiará todo a minúsculas y si hay espacios se cambiaran a guiones bajos (_).

Algunos títulos son largos, como Exercise_Induced_Angina y Heart_Attack_Risk. Se van a abreviar para facilitar su uso, pero asegurándome de que sigan siendo comprensibles.

Considero que todos los demás títulos son descriptivos y no requiere más cambios.





```python
print("Nombres de columnas originales:\n")
print(df.columns)
print()

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('exercise_induced_angina', 'ex_angina')
df.columns = df.columns.str.replace('heart_attack_risk', 'heart_risk')
df.columns = df.columns.str.replace('max_heart_rate_achieved', 'max_heart_rate')
df.columns = df.columns.str.replace('alcohol_consumption', 'alc_cons')
df.columns = df.columns.str.replace('physical_activity_level', 'phy_act_lvl')
df.columns = df.columns.str.replace('cholesterol_level', 'cholesterol_lvl')

print("Nombres de columnas normalizados:\n")
print(df.columns)

```


**Corrección de Tipos de Datos:**

Analizando cada uno de los campos del dataset:

Se mantienen como tipo int64:

*   age
*   smoking
*   alc_cons
*   diabetes
*   hypertension
*   resting_bp
*   heart_rate

Los que cambian son los que dicen tipo Object:


*   gender

Esto es porque es mejor usar el tipo category para datos categóricos con un número limitado de valores únicos, ya que esto mejora el rendimiento y reduce el uso de memoria.


```python
print("Tipos de datos originales:\n")
print(df.dtypes)
print()

df['gender'] = df['gender'].astype('category')
df['phy_act_lvl'] = df['phy_act_lvl'].astype('category')
df['stress_level'] = df['stress_level'].astype('category')
df['chest_pain_type'] = df['chest_pain_type'].astype('category')
df['thalassemia'] = df['thalassemia'].astype('category')
df['ecg_results'] = df['ecg_results'].astype('category')
df['heart_risk'] = df['heart_risk'].astype('category')

print("Tipos de datos cambiados:\n")
print(df.dtypes)
```

# **Limpieza, Transformación y Normalización de Datos Categóricos:**








Verificar valores únicos por columna categórica.

```python
for col in ['gender', 'phy_act_lvl', 'stress_level', 'chest_pain_type', 'thalassemia', 'ecg_results', 'heart_risk']:
    print(f"Valores únicos en la columna '{col}':")
    print(df[col].unique())
    print()

for col in ['gender', 'phy_act_lvl', 'stress_level', 'chest_pain_type', 'thalassemia', 'ecg_results', 'heart_risk']:
    print(f"Valores únicos en la columna '{col}':")
    print(df[col].value_counts())
    print()
```

Para limpiar los datos categoricos se va a:

1.   Eliminar  duplicados
2.   Manejar valores nulos (si hubiera)
3.   Convertir a minusculas


```python
print("DataFrame antes de la limpieza:\n")
print(df)

df = df.drop_duplicates()
df.fillna('Desconocido', inplace=True)

df['gender'] = df['gender'].str.lower()
df['phy_act_lvl'] = df['phy_act_lvl'].str.lower()
df['stress_level'] = df['stress_level'].str.lower()
df['chest_pain_type'] = df['chest_pain_type'].str.lower()
df['thalassemia'] = df['thalassemia'].str.lower()
df['ecg_results'] = df['ecg_results'].str.lower()
df['heart_risk'] = df['heart_risk'].str.lower()

print()
print("DataFrame después de la limpieza:\n")
print(df)


```

##**Outliers:**

Para revisar si el dataset contaba con outliers realice lo siguiente:

Descripción estadística de los datos - numérica: desviación estándar

Es posible detectar aquellos datos atípicos usando la desviación estándar.

Si Valori>(3×std) entonces es un valor atípico. Se usa describe para mostrar la desviación estándar por columna. Finalmente, muestro los valores que cumplen con dicha regla.
En este caso no salieron datos atípicos.


```python

print("Descripción estadística de cada una de las columnas:\n")
descripcion = df.describe()
print(descripcion)


numerical_df = df.select_dtypes(include=np.number)


print('\nDesviación estándar por columna ')
print(numerical_df.std())


atipicos = df[(np.abs(numerical_df - numerical_df.mean()) > (3 * numerical_df.std())).any(axis=1)]

print('\nValores atípicos por columna ')
print(atipicos)
```
