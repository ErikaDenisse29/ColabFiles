**Dataset:**
En este caso, vamos a utilizar el dataset de "tips" de seaborn, que contiene información sobre las propinas que dejaron los clientes en un restaurante, junto con detalles sobre el total de la cuenta, el día de la semana, si fue almuerzo o cena, entre otros.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
```

**Carga y exploración del dataset:**

Cargar el dataset tips usando seaborn.
Realizar un análisis preliminar: utiliza .head(), .info(), .describe() para obtener una visión general de los datos.
Verificar la existencia de valores nulos o datos faltantes.

```python
print(df.head())
print(df.info())
print(df.describe())
```

**a) Mapa de Calor (Heatmap):**

Objetivo: Visualizar la correlación entre las variables numéricas en el dataset.
Instrucciones: Utiliza un mapa de calor para mostrar la matriz de correlación entre las variables numéricas (total_bill, tip, size).
Asegúrate de personalizar el mapa con una paleta de colores y etiquetas en los ejes.

```python
matriz_correlacion = df[['total_bill', 'tip', 'size']].corr()

sns.heatmap(matriz_correlacion,annot=True,cmap='coolwarm')
plt.title('Matriz de correlacion')
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.show()
```

**b) Diagrama de Violín**
Objetivo: Mostrar la distribución de las propinas para diferentes días de la semana.

```python
sns.violinplot(x='day', y='tip', data=df)
plt.title('Distribucion de propinas por dia')
plt.xlabel('Dia')
plt.ylabel('Propinas')
plt.show()
```

**c) Gráfico de Dispersión**
Objetivo: Analizar la relación entre el total de la cuenta y la propina.

```python
sns.scatterplot(x='total_bill', y='tip', data=df,hue='day',palette='deep')
plt.title('Relacion entre el total de la cuenta y la propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

**d) Gráfico de Barras**
Objetivo: Comparar el promedio de la propina por género.

```python
prom_tip=df.groupby('sex')['tip'].mean()
sns.barplot(x=prom_tip.index, y=prom_tip.values, palette='pastel')
plt.title('Promedio de la propina por genero')
plt.xlabel('Genero')
plt.ylabel('Promedio de la propina')
plt.show()
```
