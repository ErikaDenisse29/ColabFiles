```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de los parmetros
N_total = 100
t_0 = 2
c = 0.1

# Clculo del valor ptimo de B
B_optimal = (2**(2/3) * N_total**(1/3) * t_0**(1/3)) / (2 * c**(1/3))

# Definicin de N(B) y t(B)
def N(B):
    return N_total / B

def t(B):
    return t_0 + (c / N_total) * B**3

# Definicin de T(B)
def T(B):
    return N(B) * t(B)

# Rango de valores para B (evitar B=0 para no dividir por cero)
B_values = np.linspace(0.1, 10, 400)

# Clculo de T(B)
T_values = T(B_values)

# Evaluar T(B) en el valor ptimo
T_optimal = T(B_optimal)

# Creacin de la figura
plt.figure(figsize=(10, 6))

# Grfico de T(B)
plt.plot(B_values, T_values, label='T(B) = N(B) * t(B)', color='blue')
plt.scatter(B_optimal, T_optimal, color='red')  # Punto ptimo
plt.text(B_optimal, T_optimal + 10, f'ptimo en ({B_optimal:.2f}, {T_optimal:.2f})', fontsize=10, color='red')
plt.title('Grfico de T(B) = N(B) * t(B) con Valor ptimo')
plt.xlabel('B (Parmetro)')
plt.ylabel('T(B) (Tiempo Total de Entrenamiento en segundos)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()

# Mostrar grfico
plt.show()

# Imprimir el valor ptimo de B y T(B)
print(f'Valor ptimo de B: {B_optimal:.2f}')
print(f'Tiempo total de entrenamiento en el valor ptimo: {T_optimal:.2f}')
```
