```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")
print (titanic)
```

Visualizar distribucion de las edades de los pasajeros

```python
sns.histplot(titanic['age'], kde=True, color='blue',bins=50)
plt.title('Distribucion de edades de los pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.barplot(x='sex',y='survived',data=titanic,palette='pastel')
plt.title('Tasa de supervivencia')
plt.xlabel('Genero')
plt.ylabel('Supervivencia')
plt.show()
```

```python
sns.catplot(x='pclass',hue='survived',kind='count',data=titanic,palette='Set2')
plt.title('Supervivencia por clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de pasajeros')
plt.show()
```

```python
sns.boxplot(x='pclass',y='age',data=titanic,palette='coolwarm')
plt.title('Distribucion de edades por clase de los pasajeros')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.show()
```

Crear matriz de correlacion

```python
num_col=titanic.select_dtypes(include=['number']).columns
corr=titanic[num_col].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.4f',linewidths=0.5)
plt.title('Matriz de correlacion')
plt.show()
```

```python
sns.pairplot(titanic[['age','fare','pclass','survived']])
plt.suptitle('Pairplot de las variables del titanic')
plt.show()
```

grafica de violin

```python
sns.violinplot(x='pclass',y='age',data=titanic,palette='muted')
plt.title('Distribucion de edades de los pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic,hue='survived',palette='coolwarm')
plt.title('Relacion entre edad y tarifa de supervivencia')
plt.xlabel('Edad')
plt.ylabel('clase')
plt.show()
```
