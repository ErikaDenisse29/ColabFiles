Ejercicio 1: Normalización Min-Max

1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, alcohol).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Min-Max a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1. ¿Cómo afectó la normalización Min-Max a la distribución de los datos?

A simple vista de los graficos no veo una diferencia significativa al hacer la transformacion.

2. ¿Qué interpretas del p-valor antes y después de la normalización?

La normalización Min-Max, en este caso, no ha logrado transformar los datos de manera que se asemejen a una distribución normal.
Por lo que hay que considerar otras transformaciones.
La falta de normalidad en los datos puede afectar la validez de los resultados de los modelos estadísticos.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
from scipy import stats


wine_data = load_wine(as_frame=True)
wine_df = wine_data.frame

alcohol_data = wine_df['alcohol']

shapiro_test_original = stats.shapiro(alcohol_data)
print("Shapiro-Wilk Test (Original Data):", shapiro_test_original)

scaler = MinMaxScaler()
alcohol_normalized = scaler.fit_transform(alcohol_data.values.reshape(-1, 1))

shapiro_test_normalized = stats.shapiro(alcohol_normalized)
print("Shapiro-Wilk Test (Normalized Data):", shapiro_test_normalized)


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.histplot(alcohol_data, kde=True, color='blue', bins=15)
plt.title('Distribución de Alcohol (Original)')
plt.xlabel('Alcohol')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sns.histplot(alcohol_normalized, kde=True, color='green', bins=15)
plt.title('Distribución de Alcohol (Normalizado)')
plt.xlabel('Alcohol Normalizado')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

# Ejercicio 2: Normalización Z-Score

Instrucciones:
1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, malic_acid).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Z-Score a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1.	¿Qué diferencia observas entre la distribución original y la normalizada con Z-Score?

A simple vista de los graficos no veo una diferencia significativa al hacer la transformacion.

2.	¿Qué implicaciones tiene la normalización Z-Score en el análisis de los datos?

La normalización Z-Score, en este caso, no ha logrado transformar los datos de manera que se asemejen a una distribución normal.
Por lo que hay que considerar otras transformaciones.
La falta de normalidad en los datos puede afectar la validez de los resultados de los modelos estadísticos.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from scipy import stats

wine_data = load_wine(as_frame=True)
wine_df = wine_data.frame


malic_acid_data = wine_df['malic_acid']


shapiro_test_original = stats.shapiro(malic_acid_data)
print("Shapiro-Wilk Test (Original Data):", shapiro_test_original)

scaler = StandardScaler()
malic_acid_normalized = scaler.fit_transform(malic_acid_data.values.reshape(-1, 1))


shapiro_test_normalized = stats.shapiro(malic_acid_normalized)
print("Shapiro-Wilk Test (Normalized Data):", shapiro_test_normalized)


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.histplot(malic_acid_data, kde=True, color='blue', bins=15)
plt.title('Distribución de Ácido Malico (Original)')
plt.xlabel('Ácido Malico')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sns.histplot(malic_acid_normalized, kde=True, color='green', bins=15)
plt.title('Distribución de Ácido Malico (Normalizado)')
plt.xlabel('Ácido Malico Normalizado')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Ejercicio 3: Transformación Logarítmica
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación logarítmica (np.log) a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación logarítmica a la distribución de los datos?

La transformación logarítmica ayuda a reducir la asimetría y estabilizar la varianza. Al ver los graficos si se nota la diferencia y se ve mas distrubuida la informacion.

2.	¿Qué interpretas del p-valor antes y después de la transformación logarítmica?

Puedo ver aumento el p-valor después de la transformación logarítmica lo cual indica que los datos transformados son más compatibles con la normalidad.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


np.random.seed(42)
data_original = np.random.exponential(scale=1.0, size=1000)  # Datos aleatorios


data_transformed = np.log(data_original)


shapiro_test_original = stats.shapiro(data_original)
shapiro_test_transformed = stats.shapiro(data_transformed)

print("Shapiro-Wilk Test (Original Data):", shapiro_test_original)
print("Shapiro-Wilk Test (Transformed Data):", shapiro_test_transformed)


plt.figure(figsize=(12, 6))

# Gráfico de datos originales
plt.subplot(1, 2, 1)
sns.histplot(data_original, kde=True, color='blue', bins=30)
plt.title('Distribución de Datos Originales')
plt.xlabel('Datos Originales')
plt.ylabel('Frecuencia')

# Gráfico de datos transformados
plt.subplot(1, 2, 2)
sns.histplot(data_transformed, kde=True, color='green', bins=30)
plt.title('Distribución de Datos Transformados (Logarítmica)')
plt.xlabel('Datos Transformados')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Ejercicio 4: Transformación Raíz Cuadrada
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación de raíz cuadrada a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación raíz cuadrada a la distribución de los datos?

Comparando las graficas, despues de aplicar la tranformacion la grafica luce  más a una distribución normal. Los datos se ven más uniformes. Ya no se ven tantos valores extremos
Esto ayuda a tener una mejor interpretacion de los datos.

2.	¿Qué efectos tiene la raíz cuadrada sobre los valores extremos de los datos?

Al parecer reduce la magnitud de los valores extremos lo cual ayuda a
tener una mejor representación de la tendencia central de los datos.
Pero esto podria implicar a una pérdida de información original de los datos, afectando el resultado.




```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


np.random.seed(42)
data_original = np.random.exponential(scale=1.0, size=1000)  # Datos aleatorios


data_transformed = np.sqrt(data_original)


shapiro_test_original = stats.shapiro(data_original)
shapiro_test_transformed = stats.shapiro(data_transformed)

print("Shapiro-Wilk Test (Original Data):", shapiro_test_original)
print("Shapiro-Wilk Test (Transformed Data):", shapiro_test_transformed)


plt.figure(figsize=(12, 6))

# Gráfico de datos originales
plt.subplot(1, 2, 1)
sns.histplot(data_original, kde=True, color='blue', bins=30)
plt.title('Distribución de Datos Originales')
plt.xlabel('Datos Originales')
plt.ylabel('Frecuencia')

# Gráfico de datos transformados
plt.subplot(1, 2, 2)
sns.histplot(data_transformed, kde=True, color='green', bins=30)
plt.title('Distribución de Datos Transformados (Raíz Cuadrada)')
plt.xlabel('Datos Transformados')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Ejercicio 5: Transformación Box-Cox
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar la transformación Box-Cox a los datos utilizando scipy.stats.boxcox.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación Box-Cox a la distribución de los datos?

Hizo que la distribución de los datos se asemeje más a una distribución normal
Al comparar las gráficas, la distribución transformada tiene una forma más simétrica en comparación con la distribución original.

2.	¿Qué interpretación tiene el valor lambda obtenido en la transformación Box-Cox?

El valor de lambda es positivo y menor que 1. Esto significa que la transformación está moderadamente ajustada para estabilizar la varianza y hacer que la distribución sea más normal.
Tambien si es menor que 1 sugiere que la transformación ayuda a reducir la asimetría de la distribución y también ayuda a mitigar el impacto de los valores extremos, acercándolos más a la media de la distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


np.random.seed(42)
data_original = np.random.exponential(scale=1.0, size=1000)  # Datos aleatorios


data_transformed, lambda_value = stats.boxcox(data_original)


shapiro_test_original = stats.shapiro(data_original)
shapiro_test_transformed = stats.shapiro(data_transformed)

print("Shapiro-Wilk Test (Original Data):", shapiro_test_original)
print("Shapiro-Wilk Test (Transformed Data):", shapiro_test_transformed)
print("Valor de lambda en la transformación Box-Cox:", lambda_value)


plt.figure(figsize=(12, 6))

# Gráfico de datos originales
plt.subplot(1, 2, 1)
sns.histplot(data_original, kde=True, color='blue', bins=30)
plt.title('Distribución de Datos Originales')
plt.xlabel('Datos Originales')
plt.ylabel('Frecuencia')

# Gráfico de datos transformados
plt.subplot(1, 2, 2)
sns.histplot(data_transformed, kde=True, color='green', bins=30)
plt.title('Distribución de Datos Transformados (Box-Cox)')
plt.xlabel('Datos Transformados')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
